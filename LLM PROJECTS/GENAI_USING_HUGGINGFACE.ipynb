```python
from langchain import PromptTemplate, LLMChain

Qs="What do you mean by recession period?"
template="""Question: {qs}
Answer: Let's think step by step."""
prompt = PromptTemplate(template=template, input_variables=["qs"])
print (prompt)
```

```python
## HUGGING PIPELINE
```

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python
model_id="gpt2"
model=AutoModelForCausalLM.from_pretrained(model_id)
tokenizer=AutoTokenizer.from_pretrained(model_id)
```

```python
pipe=pipeline("text-generation",model=model,tokenizer=tokenizer,max_new_tokens=100)
hf=HuggingFacePipeline(pipeline=pipe)
```

```python
hf
```

```python
hf.invoke("apple is a company")
```

```python
## GPU ENABLED HUGGINGFACE PIPELINE
gpu_llm=HuggingFacePipeline(pipeline=pipe)
```

```python
template="""Question: {qs}
Answer: Let's think step by step."""
prompt=PromptTemplate.from_template(template)
```

```python
chain=prompt|gpu_llm
```

```python
Qs="what is Artificial Intelligence??"
chain.invoke({"qs":Qs})
```
